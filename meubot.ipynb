```python
df = pd.DataFrame(solicitacoes)
df.coluns = [ "Nome_Solicitacao", "Numero",
              "Data_Abertura","Categoria",
              "Projeto","Prioridade",
              "Descricaoo","Agendamentos",
              "Requisicoes","Status" ]
df
```

Agora preciso 'embedar' as linhas da tabela para poder fazer o calculo escalar dos vetores para a análise de contexto

para isso, concatenei todos os itens importantes da tabela

```python
# Selecionando as colunas relevantes
colunas_relevantes = ['Nome-Solicitação', 'Categoria',
                      'Projeto', 'Descrição',
                      "Data-Abertura","Categoria",
                      "Projeto","Prioridade",
                      "Descrição","Agendamentos",
                      "Requisições","Status" ]

# Combinando as colunas selecionadas
df['texto_combinado'] = df[colunas_relevantes].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df
```

```python
model = 'models/embedding-001'
def embed_fn(title, text):
  return genai.embed_content(model=model,
                              content=text,
                              task_type="retrieval_document",
                              title=title)['embedding']

df['embeddings'] = df.apply(lambda row: embed_fn(row['Nome-Solicitação'], row['texto_combinado']), axis=1)
df
```

```python
import numpy as np

def gerar_e_buscar_consult(consulta, base, model):
  embadding_da_consulta = genai.embed_content(model=model,
                                content=consulta,
                                task_type="retrieval_query")['embedding']

  produtos_escalares = np.dot(np.stack(df['embeddings']), embadding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]
```

```python

consulta = str(input('digite o nome do projeto: '))

gerar_e_buscar_consult(consulta, df, model)


model2 = genai.GenerativeModel(model_name='gemini-1.0-pro',
                            generation_config=generate_config,
                            safety_settings=safety_settings)

chat = model2.start_chat(history=[])


texto = gerar_e_buscar_consult(consulta, df, model)

print('--'*50)
print('\n')
print(texto)
print('\n')
print('--'*50)

consulta2 = f'Através dessas informações: {texto}  faça uma descrição do projeto para um usuário. falando de todos os tópicos em formato de tabela, os tópicos são: "Nome-Solicitação", "Categoria","Projeto", "Descrição","Data-Abertura","Categoria","Projeto","Prioridade", "Descrição","Agendamentos" e "Requisições","Status"'
response = model2.generate_content(consulta2)
print(response.text)
print('\n')
print('--'*50)

consulta3 = ''
while consulta3 != 'fim':

  consulta3 = str(input('Alguma dúvida?: '))
  requestia = f'Através dessas informações: {response}  responda essa pergunta: {consulta3}'
  response = chat.send_message(requestia)
  print(f'CHAT BOT:\n\n{response.text}')
  print('\n')
  print('--'*50)

```

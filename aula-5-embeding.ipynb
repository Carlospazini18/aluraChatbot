* TILIZANDO O DATA-FRAME COM PANDAS

```python
df = pd.DataFrame(documents)
df.columns = ['Title', 'Text']
df
```

* FTONANDO SUSTENTÁVEL: função para 'embedar' uma base de dados

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                content=sample_text,
                                task_type="retrieval_document",
                                title=title)['embedding']
```

* RODANDO A FUNÇÃO LINHA A LINHA

```python
df['embeddings'] = df.apply(lambda row: embed_fn(row['Title'], row['Text']), axis=1)
df
```

* FAZENDO A CONSULTA E A COMPRAÇÃO VETORIAL DOS EMBEDDINGS

```python
def gerar_e_buscar_consult(consulta, base, model):
  embadding_da_consulta = genai.embed_content(model=model,
                                content=consulta,
                                task_type="retrieval_query")['embedding']

  produtos_escalares = np.dot(np.stack(df['embeddings']), embadding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]['Text']
```

* TEXTANDO O MODELO

```python
consulta = "How do you shift gears in the Google car?"
model = 'models/embedding-001'

gerar_e_buscar_consult(consulta, df, model)
```
